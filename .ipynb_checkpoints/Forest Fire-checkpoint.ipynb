```python
import pandas as pd

# Group the DataFrame by the column with duplicated values and check if other variables are same
duplicated_grouped = forest_fires[forest_fires.duplicated(['Fire ID'], keep=False)].groupby('Fire ID')

# Loop through each group and drop rows based on variable values
for group_name, group_data in duplicated_grouped:
    if len(group_data.groupby(group_data.columns.tolist())) > 1:
        # Drop all rows if variable values are different
        forest_fires = forest_fires.drop(group_data.index)
    else:
        # Drop one row if variable values are the same
        forest_fires = forest_fires.drop(group_data.index[0])
```

## Converting csv files from Scraping to single dataframe

```python
import os
import pandas as pd

folder_path = 'factors csv'  # Replace with your folder path
col_names = ["time", "weathercode (wmo code)", "temperature_2m_max (Â°C)", "temperature_2m_min (Â°C)", "temperature_2m_mean (Â°C)", "apparent_temperature_max (Â°C)", "apparent_temperature_min (Â°C)", "apparent_temperature_mean (Â°C)", "sunrise (iso8601)", "sunset (iso8601)", "shortwave_radiation_sum (MJ/mÂ²)", "precipitation_sum (mm)", "rain_sum (mm)", "snowfall_sum (cm)", "precipitation_hours (h)", "windspeed_10m_max (km/h)", "windgusts_10m_max (km/h)", "winddirection_10m_dominant (Â°)", "et0_fao_evapotranspiration (mm)"]
df_all = pd.DataFrame(columns=col_names)

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, delimiter='\t', skiprows=[0], names=col_names)
        df_new_cols = pd.DataFrame(df['time'].str.split(',').tolist(), columns=col_names)
        df_new_cols = df_new_cols.iloc[2:]
        df_new_cols["fire_id"] = filename.split(".")[0]
        df_all = pd.concat([df_all, df_new_cols], ignore_index=True)

# The final dataframe with data from all csv files in the folder
df_all.head()
```

```python
df_all.shape
```

```python
# Check if column A has all unique values
if not df_all['fire_id'].duplicated().any():
    print("All values in column A are unique!")
else:
    print("Column A has duplicate values." + " Count " + str(df_all['fire_id'].duplicated().sum()))
```
