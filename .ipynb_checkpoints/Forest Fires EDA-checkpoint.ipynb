The dataset consists of 517 observations and 13 features.

The various features are described below.
1. X - x-axis spatial coordinate within the Montesinho park map: 1 to 9
2. Y - y-axis spatial coordinate within the Montesinho park map: 2 to 9
3. month - month of the year: 'jan' to 'dec'
4. day - day of the week: 'mon' to 'sun'
5. FFMC - FFMC index from the FWI system: 18.7 to 96.20
6. DMC - DMC index from the FWI system: 1.1 to 291.3
7. DC - DC index from the FWI system: 7.9 to 860.6
8. ISI - ISI index from the FWI system: 0.0 to 56.10
9. temp - temperature in Celsius degrees: 2.2 to 33.30
10. RH - relative humidity in %: 15.0 to 100
11. wind - wind speed in km/h: 0.40 to 9.40
12. rain - outside rain in mm/m2 : 0.0 to 6.4
13. area - the burned area of the forest (in ha): 0.00 to 1090.84

##### 2D - Data Cleaning

There are no null data but values for area is 0 in some observations. This needs to be handled in some way. 

As its our target variable, and is important 

1. We can drop those rows

Else, 

As our dataset is very small

2. We cam impute values as its all based on specific region, assuming these are missing values

For our understanding of the data better, lets go ahead with both options

```python
df_dropped_area = forest_fires.drop(forest_fires[forest_fires['area'] == 0].index)
df_dropped_area.info()
```

We will go ahead with Multiple Imputation - https://scikit-learn.org/stable/modules/impute.html

This method is computationaly intensive, but our dataset is small hence it is feasible. Also it tries different imputation methods and provides us with the one that is more statistically accurate. 

Multiple imputation accounts for the uncertainty associated with imputed values and can provide more accurate estimates of parameters and uncertainties in the analysis

```python
dataset_imputed_area = forest_fires.copy()

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Replace 0 values with NaN
dataset_imputed_area['area'] = dataset_imputed_area['area'].replace(0, float('nan'))

# Perform multiple imputation using IterativeImputer
imputer = IterativeImputer(n_nearest_features=5, initial_strategy='median', random_state=0)
dataset_imputed_area['area'] = imputer.fit_transform(dataset_imputed_area[['area']])
```

### Step 4 - EDA

##### 4A - Univariate Analysis

1. Lets see variable FFMC

```python
def univariate_analysys(target_name, target):
    # Histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(target, bins=10)
    plt.title(target_name)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()

    # Box plot
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=target)
    plt.title(target_name)
    plt.ylabel("Value")
    plt.show()

    # Measures of central tendency
    print("Mean:", target.mean())
    print("Median:", target.median())
    print("Mode:", target.mode().iloc[0])

    # Measures of dispersion
    print("Range:", target.max() - df['FFMC'].min())
    print("Variance:", target.var())
    print("Standard Deviation:", target.std())

    # Measure of skewness
    print("Measure of Skewness:", target.skew())
```

```python
univariate_analysys('FFMC', dataset_imputed_area['FFMC'])
```

Data is left skewed. Hence we can apply exponential transformation

```python
from scipy.stats import boxcox
lambda_value = 16
new_target = boxcox(dataset_imputed_area['FFMC'], lmbda=lambda_value)
univariate_analysys('Exponential FFMC', new_target)
```

##### 4B - Bivariate Analysis

Lets setup data for bivariant analysis from previous analysis

```python
data_bivariant = dataset_imputed_area.copy()
lambda_value = 16
data_bivariant['FFMC'] = boxcox(data_bivariant['FFMC'], lmbda=lambda_value)
lambda_value = 4
data_bivariant['DC'] = boxcox(data_bivariant['DC'], lmbda=lambda_value)
data_bivariant.drop(data_bivariant[data_bivariant['DC'] == 0].index, inplace=True)
data_bivariant['ISI'] = np.sqrt(data_bivariant['ISI'])
data_bivariant["temp"] = np.sqrt(data_bivariant["temp"])
data_bivariant['area'] = np.log(dataset_imputed_area['area'])
```

```python
# Create a list of variables
variables = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']

# Create a pairplot
sns.pairplot(data_bivariant[variables])
plt.suptitle('Pairplot of Variables')
plt.show()
```
